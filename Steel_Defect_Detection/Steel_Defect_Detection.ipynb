```python
import cv2
import numpy as np
import random
from matplotlib import pyplot as plt

img1 = cv2.imread('D:/dataset/severstal-steel-defect-detection/train_images/000a4bcdd.jpg',0)
img2 = cv2.imread('D:/dataset/severstal-steel-defect-detection/train_images/000f6bf48.jpg',0)

hist1 = cv2.calcHist([img1],[0],None,[256],[0,256])
hist2 = cv2.calcHist([img2],[0],None,[256],[0,256])

plt.subplot(221),plt.imshow(img1,'gray'),plt.title('Red Line')
plt.subplot(222),plt.imshow(img2,'gray'),plt.title('Green Line')
plt.subplot(223),plt.plot(hist1,color='r'),plt.plot(hist2,color='g')

plt.xlim([0,256])
plt.show()
```
